# QUESTION 5: PLOT A GRAPH TO SHOW THE HIGHEST AND LOWEST SCORES IN THE READING TEST IN DESCENDING ORDER

```python
fig = plt.figure(figsize = (20,30))
df_sorted_desc= df_students.sort_values('reading score',ascending=False)
plt.plot(df_students['reading score'], data = df_sorted_desc,color = 'red', ls = '--', lw = 0.2, marker = 'o',
         markerfacecolor = 'k', markeredgecolor = 'y')
plt.title('HIGHEST TO LOWEST SCORE OF READING TEST IN DESCENDING ORDER', color = 'g', fontsize = 25)
plt.show()
```

# The graph shows that the highest score in the reading test was 100, while the lowest was 17

# QUESTION 6: PLOT A GRAPH SHOWING THE HIGHEST AND LOWEST SCORES IN THE WRITING TEST 

```python
fig = plt.figure(figsize = (20,30))
df_sorted_desc= df_students.sort_values('writing score',ascending=False)
plt.plot(df_students['writing score'], data = df_sorted_desc,color = 'g', ls = '--', lw = 0.2, marker = 'h',
         markerfacecolor = 'r', markeredgecolor = 'y')
plt.title('HIGHEST TO LOWEST SCORE OF WRITING TEST IN DESCENDING ORDER', color = 'r', fontsize = 25)
plt.show()
```

 # The graph shows that the highest score obtained for the writing test is 100, while the lowest score is 10

# QUESTION 8: CONVERTING DATA FRAME TO EXCEL WITH EACH GROUP IN A DIFFERENT SHEET

```python
group_a = df_students[df_students['race/ethnicity'] == 'group A']
group_b = df_students[df_students['race/ethnicity'] == 'group B']
group_c = df_students[df_students['race/ethnicity'] == 'group C']
group_d = df_students[df_students['race/ethnicity'] == 'group D']
group_e = df_students[df_students['race/ethnicity'] == 'group E']
```

```python
writer = pd.ExcelWriter('output.xlsx')
group_a.to_excel(writer, 'GROUP A')
group_b.to_excel(writer, 'GROUP B')
group_c.to_excel(writer, 'GROUP C')
group_d.to_excel(writer, 'GROUP D')
group_e.to_excel(writer, 'GROUP E')
writer.save()
```
