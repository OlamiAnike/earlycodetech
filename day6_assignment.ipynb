# QUESTION ONE
# TOP FIVE INDUSTRIES FROM 2000 - 2009

```python
year = gdp2['tyear']
agric =gdp2['Agriculture'] #highest
oil = gdp2['OilandGas'] #second highest
estate = gdp2['RealEstate'] #third
trade = gdp2['Trade'] #fourth
manu = gdp2['Manufacturing'] #fifth
```

```python
fig = plt.figure(figsize = (25,15))
ax = fig.add_axes([0.1,0.1,0.6,0.6])
ax.plot(year, agric,'red', ls = '--', lw = 2, marker = 'o', markerfacecolor = 'b', markeredgecolor = 'y', label = 'Agriculture')
ax.plot(year, oil, 'yellow', ls = '--', lw = 2, marker = 'h', markerfacecolor = 'k', markeredgecolor = 'b', label = 'OilandGas')
ax.plot(year, trade, 'red', ls = '--', lw = 2, marker = 'd', markerfacecolor = 'y', markeredgecolor = 'g', label = 'Trade')
ax.plot(year, estate, 'black', ls = '--', lw = 2, marker = 'h', markerfacecolor = 'g', markeredgecolor = 'b', label = 'Real Estate')
ax.plot(year, manu, 'blue', ls = '--', lw = 2, marker = 'o', markerfacecolor = 'r', markeredgecolor = 'k', label = 'Manufacturing')
ax.set_xlabel('YEAR', color = 'k', fontsize = 20)
ax.set_ylabel('INDUSTRIES', color = 'r', fontsize = 20)
ax.set_title("TOP FIVE CONTRIBUTING INDUSTRIES TO NIGERIA'S GDP FROM 2000- 2009", color = 'b', fontsize = 25)
ax.legend()
plt.show()
```

# REASONS FOR GROWTH

# QUESTION 2
# LEAST FIVE INDUSTRIES FROM 2000 - 2009

```python
year = gdp2['tyear']
util =gdp2['Utilities']  #fifth
food = gdp2['Food/services'] #fourth lowest
forestry = gdp2['Forestry'] # third lowest
solid = gdp2['SolidMinerals'] #second lowest
admin = gdp2['Admin-support'] #lowest
```

```python
fig = plt.figure(figsize = (25,15))
ax = fig.add_axes([0.1,0.1,0.6,0.6])
ax.plot(year, util,'red', ls = '--', lw = 2, marker = 'o', markerfacecolor = 'b', markeredgecolor = 'y', label = 'Utilities')
ax.plot(year, food, 'yellow', ls = '--', lw = 2, marker = 'h', markerfacecolor = 'k', markeredgecolor = 'b', label = 'Food/services')
ax.plot(year, forestry, 'red', ls = '--', lw = 2, marker = 'd', markerfacecolor = 'y', markeredgecolor = 'g', label = 'Forestry')
ax.plot(year, solid, 'black', ls = '--', lw = 2, marker = 'h', markerfacecolor = 'g', markeredgecolor = 'b', label = 'Solid Minerals')
ax.plot(year, admin, 'blue', ls = '--', lw = 2, marker = 'o', markerfacecolor = 'r', markeredgecolor = 'k', label = 'Admin Support')
ax.set_xlabel('YEAR', color = 'b', fontsize = 20)
ax.set_ylabel('INDUSTRIES', color = 'purple', fontsize = 20)
ax.set_title("LEAST FIVE CONTRIBUTING INDUSTRIES TO NIGERIA'S GDP FROM 2000- 2009", color = 'r', fontsize = 25)
ax.legend()
plt.show()
```

# SOLUTIONS
# The least five industries showed a reasonable amount of increase except for admin support which was more or less stagnant. 
# A suitable solution would be to put in more resources into growing these industries and if constant support is given, there will be increase over time.
